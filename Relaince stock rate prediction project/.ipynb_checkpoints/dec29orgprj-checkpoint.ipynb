```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('reliance_data.csv')     #reading the csv file 
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns = ['Symbol','Series','Prev Close','VWAP','Trades','Deliverable Volume','%Deliverble'],axis=1)     #droping unnecessary columns  
df
```

```python
df.isnull().sum()   #finding total no of null values in each column
```

```python
df['Date'] = pd.to_datetime(df.Date,format='%d-%m-%Y')
df.index = df['Date']
plt.figure(figsize=(16,8))
plt.plot(df['Close'], label='Close Price history')
```

```python
a = df['Last'].mean()
a
```

```python
df['Last'] = df['Last'].fillna(a)
df
```

```python
df.shape
```

```python
df_x = df.drop(columns = ['Date','Close'],axis=1)   #droping the output columns to start training and testing 
df_y = df[['Date','Close']]
print(df_x)
print(df_y)
```

```python
y_train = df_y[:4964]
y_test = df_y[4964:]
print(y_train.shape)
print(y_test.shape)
```

```python
preds = []
for i in range(0,y_test.shape[0]):
    x = y_train['Close'][len(y_train)-1241+i:].sum() + sum(preds)
    y = x/1241
    preds.append(y)
```

```python
rms=np.sqrt(np.mean(np.power((np.array(y_test['Close'])-preds),2)))
print('\n RMSE value on validation set:')
print(rms)
```

```python
y_test['Predictions'] = 0
y_test['Predictions'] = preds
plt.plot(y_train['Close'])
plt.plot(y_test[['Close', 'Predictions']])
```
